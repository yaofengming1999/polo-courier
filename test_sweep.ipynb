```python
from math import prod
from pathlib import Path

import pandas as pd
import wandb
import yaml

PROJECT = "minicourier-new-cikm3"
SWEEP_DIR = Path("yamls/sweeps/sweep_greedy")
RUN_TO_BASE = {
    "run_tower.py": Path("yamls/base_actower.yaml"),
    "run_tcac.py": Path("yamls/base_tcac.yaml"),
    "run_greedy.py": Path("yamls/base_greedy.yaml"),
}


def extract_instance_size(sweep_config):
    values = (
        sweep_config.get("parameters", {})
        .get("instance_size", {})
        .get("values", [])
    )
    if not values:
        return None
    return values[0] if len(values) == 1 else values


def load_base_config(sweep_config):
    program = sweep_config.get("program")
    base_path = RUN_TO_BASE.get(program)
    if base_path is None:
        return {}
    with base_path.open() as f:
        return yaml.safe_load(f)


def extract_swept_parameters(sweep_config, base_config):
    swept = []
    parameters = sweep_config.get("parameters", {})
    for name, spec in parameters.items():
        values = spec.get("values", [])
        if len(values) > 1:
            swept.append(name)
        elif len(values) == 1:
            base_value = base_config.get(name)
            only_value = values[0]
            if isinstance(only_value, bool) and base_value != only_value:
                swept.append(name)
    return swept


def extract_run_count(sweep_config):
    parameters = sweep_config.get("parameters", {})
    counts = []
    for spec in parameters.values():
        values = spec.get("values", [])
        counts.append(len(values) if values else 1)
    return prod(counts) if counts else 0


sweep_records = []

for sweep_path in sorted(SWEEP_DIR.glob("*.yaml")):
    with sweep_path.open() as f:
        sweep_config = yaml.safe_load(f)
    base_config = load_base_config(sweep_config)

    sweep_id = wandb.sweep(sweep_config, project=PROJECT)
    instance_size = extract_instance_size(sweep_config)
    swept_parameters = extract_swept_parameters(sweep_config, base_config)
    run_count = extract_run_count(sweep_config)

    record = {
        "file": sweep_path.name,
        "sweep_id": sweep_id,
        "instance_size": instance_size,
        "swept_parameters": swept_parameters,
        "run_count": run_count,
    }
    sweep_records.append(record)
    print(
        f"{record['file']}: sweep_id={record['sweep_id']}, "
        f"instance_size={record['instance_size']}, "
        f"swept_parameters={record['swept_parameters']}, "
        f"run_count={record['run_count']}"
    )

sweep_table = pd.DataFrame(sweep_records)
sweep_table["swept_parameters"] = sweep_table["swept_parameters"].apply(
    lambda xs: ", ".join(xs)
)

display(sweep_table)
sweep_records
```
